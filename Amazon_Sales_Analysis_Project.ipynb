# **Exploratory Data Analysis**

Analysis the Data

**Item** **Type**

```python
#plotting the Between item type and count of item type
ax=sns.countplot(x='Item Type',data=df)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
```

From above graph we can see that the Item like Office Supplies,fruits,Personal Care, Clothes,Cosmetics and Household are sold in large quantities whereas Meat and Snacks are sold very less.

```python
#plotting the Between item type and Profit of item type
sales_item=df.groupby(['Item Type'],as_index=False)['Total Profit'].sum().sort_values(by='Total Profit',ascending=False).head(10)
ax = sns.barplot(x='Item Type', y='Total Profit', data=sales_item)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
```

From above graph we can see that the Profit is very high on  Cosmetics,Household and Office Supplies whereas Profit on  Personal Care,Beverages and Snacks is very less.

**Region**

```python
#plotting the Region
ax=sns.countplot(x='Region',data=df)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
```

From above graph we can see that Europe and Sub-Saharan Africa has very high sales while 'Central America and the Caribbean' and North America has very low sales

```python
#plotting the Total Profit by Region
sales_item=df.groupby(['Region'],as_index=False)['Total Profit'].sum().sort_values(by='Total Profit',ascending=False).head(10)
ax = sns.barplot(x='Region', y='Total Profit', data=sales_item)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
```

Profit is very High in Sub-Saharan Africa and Europe Whereas profit is very less in 'Central America and the Caribbean' and North America.

**Cost** **By** **Item**

```python
plotting the item type by unit cost
sales_item=df.groupby(['Item Type'],as_index=False)['Unit Cost'].sum().sort_values(by='Unit Cost',ascending=False).head(10)
ax = sns.barplot(x='Item Type', y='Unit Cost', data=sales_item)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
```

The unit cost of Office Supplies,Household and Cosmetics is very high whereas the unit cost of Personal Care,Vegetables,Clothes and Snacks is very low

**Country wise Total Revenue**

Top 10 Countries with high revenue

```python
#plotting the countrywise Total Revenue
country_revenue = df.groupby('Country')['Total Revenue'].sum().sort_values(ascending=False).head(10)

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(country_revenue, labels=country_revenue.index, autopct='%1.1f%%', startangle=140)
```

From above graph we can see that Honduras,Myanmar and Djibouti have the highest revenue.

Top 10 Countries with low revenue

```python
#plotting the countrywise Total Revenue
country_revenue = df.groupby('Country')['Total Revenue'].sum().sort_values(ascending=True).head(10)

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(country_revenue, labels=country_revenue.index, autopct='%1.1f%%', startangle=140)

```

From above graph we can see that South Sudan,Mali and Lesotho have the Lowest revenue.
