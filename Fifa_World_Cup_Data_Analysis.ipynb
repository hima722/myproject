# **Exploratory** **Data** **Analysis**

```python
#count of each unique value in the 'Stage' column of the DataFrame
stage_matches =WorldCupMatches1['Stage'].value_counts()

# Create the barplot for stages
plt.figure(figsize=(14, 8))  # Increase the figure size
sns.barplot(x=stage_matches.index, y=stage_matches.values, palette='viridis')
plt.title('Number of Matches Played in Each Stage')
plt.xlabel('Stage')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

From above graph we can see that the Maximum number of Matches Played on stages of Round of 16,Quarter-finals,Group 1 and Group A and the less number of matches played on stages of Group 5,Third place,Play-off for third place.

```python
# Create the barplot
plt.figure(figsize=(10, 6))
plt.hist(WorldCupMatches['Year'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Matches Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Matches')
plt.grid(True)
plt.tight_layout()
plt.show()
```

From above graph we can see that the Distribution of matches is continuously increasing with time.

```python
#containing all unique team names participating in the World Cup matches
teams = WorldCupMatches1['Home Team Name'].append(WorldCupMatches1['Away Team Name'])

# Create a bar plot for the frequency of matches involving each team
plt.figure(figsize=(14, 8))
sns.countplot(x=teams, palette='viridis', order=teams.value_counts().index)
plt.title('Frequency of Matches Involving Each Team')
plt.xlabel('Team')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

From above graph we can see that the Frequency of Matches is Highest for Brazil,Italy,Argentina and Frequency of Matches is lowest for Kuwait,Dutch East Indies

```python
# Assuming 'Attendance' is the column containing attendance data
plt.figure(figsize=(10, 6))
plt.hist(WorldCupMatches1['Attendance'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Attendance at Matches')
plt.xlabel('Attendance')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

From above graph we can see that the distribution of attendance at matches

```python
plt.figure(figsize=(12, 6))

# Plot histogram for half-time home goals
plt.hist(WorldCupMatches1['Half-time Home Goals'], bins=10, color='skyblue', alpha=0.7, label='Half-time Home Goals')
# Plot histogram for half-time away goals
plt.hist(WorldCupMatches1['Half-time Away Goals'], bins=10, color='orange', alpha=0.7, label='Half-time Away Goals')

plt.title('Distribution of Half-time Goals')
plt.xlabel('Goals')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

From the above Graph,we can analyze the distribution of half-time goals scored by home and away teams in World Cup matches. The overlapping histograms allow us to compare the frequency of different goal ranges.

```python
#Graph for Refree,Assistant 1,Assistant 2
officials = ['Referee', 'Assistant 1', 'Assistant 2']

plt.figure(figsize=(12, 8))

# Create a bar plot for each official
for official in officials:
    plt.subplot(3, 1, officials.index(official) + 1)
    sns.countplot(y=official, data=WorldCupMatches1, palette='viridis',
                  order=WorldCupMatches1[official].value_counts().index[:10])  # Limit to top 10 officials
    plt.title(f'Frequency of Matches Officiated by {official}')
    plt.xlabel('Frequency')
    plt.ylabel(official)
    plt.tight_layout()

plt.show()
```

From the above Graph,This code generates subplots displaying the frequency of matches officiated by referees, assistant referees 1, and assistant referees 2 using a bar plot for each official, limited to the top 10 officials based on their frequency of officiating matches.

```python
WorldCupMatches1['Match Pair'] = WorldCupMatches1['Home Team Initials'] + ' vs ' + WorldCupMatches1['Away Team Initials']

# Calculate the frequency of each match pair
match_pair_counts = WorldCupMatches1['Match Pair'].value_counts()

# Plotting the distribution of match pairs
plt.figure(figsize=(10, 6))
match_pair_counts.head(20).plot(kind='bar', color='skyblue')
plt.title('Top 20 Match Pairs')
plt.xlabel('Match Pair')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

This code creates a new column 'Match Pair' by combining the home team initials and away team initials in the DataFrame 'WorldCupMatches1'. Then, it calculates the frequency of each match pair and plots the distribution of the top 20 match pairs using a bar plot.
